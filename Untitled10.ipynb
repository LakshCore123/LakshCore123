```python
## Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

pd.set_option('display.float_format', '{:.4f}'.format)
print("✅ All libraries imported successfully.")
```

```python
## Step 2: Load the Dataset
try:
    data = fetch_california_housing(as_frame=True)
    df = pd.concat([data.data, data.target.rename("HousePrice")], axis=1)
    print("✅ California Housing Dataset loaded from scikit-learn.")
except Exception:
    # Fallback: reproducible synthetic dataset with same structure
    print("⚠️  Could not download dataset. Using reproducible synthetic equivalent.")
    np.random.seed(42)
    n = 20640
    MedInc     = np.random.lognormal(1.5, 0.7, n)
    HouseAge   = np.random.uniform(1, 52, n)
    AveRooms   = np.random.lognormal(1.7, 0.4, n)
    AveBedrms  = AveRooms / np.random.uniform(3, 6, n)
    Population = np.random.lognormal(6, 1, n)
    AveOccup   = np.random.lognormal(1.1, 0.3, n)
    Latitude   = np.random.uniform(32.5, 42.0, n)
    Longitude  = np.random.uniform(-124.5, -114.3, n)
    HousePrice = np.clip(
        0.45*MedInc + 0.02*HouseAge + 0.05*AveRooms
        - 0.03*AveOccup - 0.1*(Latitude - 37)
        + np.random.normal(0, 0.5, n), 0.15, 5.0
    )
    df = pd.DataFrame({
        'MedInc': MedInc, 'HouseAge': HouseAge, 'AveRooms': AveRooms,
        'AveBedrms': AveBedrms, 'Population': Population, 'AveOccup': AveOccup,
        'Latitude': Latitude, 'Longitude': Longitude, 'HousePrice': HousePrice
    })

print(f"Dataset shape: {df.shape}")
df.head()
print("Data Types:")
print(df.dtypes)
print(f"\nMissing Values: {df.isnull().sum().sum()}")
print("\nBasic Statistics:")
df.describe()
```

```python
## Step 3: Separate Features and Target Variable
##ML models require input features `X` and a target output `y` to be handled separately.
X = df.drop("HousePrice", axis=1)
y = df["HousePrice"]

print(f"Feature matrix shape : {X.shape}")
print(f"Target vector shape  : {y.shape}")
print("\nFeatures:", list(X.columns))
```

```python
## Step 4: Feature Scaling (Critical Step)
##Features exist on very different numeric ranges. Without scaling, larger-magnitude features dominate the model and performance becomes unstable.

##StandardScaler** transforms each feature to: **mean = 0, standard deviation = 1**
print("Feature ranges BEFORE scaling:")
print(X.describe().loc[['min','max']].T.to_string())

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print("\nFeature stats AFTER scaling (mean ≈ 0, std ≈ 1):")
print(X_scaled_df.describe().loc[['mean','std']].T.to_string())
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(df['MedInc'], bins=50, color='steelblue', edgecolor='white')
axes[0].set_title('MedInc — Before Scaling', fontweight='bold')
axes[0].set_xlabel('Value'); axes[0].set_ylabel('Frequency')

axes[1].hist(X_scaled_df['MedInc'], bins=50, color='darkorange', edgecolor='white')
axes[1].set_title('MedInc — After Scaling', fontweight='bold')
axes[1].set_xlabel('Standardized Value')

plt.suptitle('Effect of StandardScaler on Feature Distribution', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('feature_scaling.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Feature scaling complete.")
```

```python
## Step 5: Train–Test Split
##We hold out 20% of data for testing. This ensures the model is evaluated on **unseen data**, not memorized values.
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
print(f"Training samples : {X_train.shape[0]}")
print(f"Testing samples  : {X_test.shape[0]}")
print(f"Features         : {X_train.shape[1]}")
```

```python
## Step 6: Train Multiple Models
##We train three algorithms:
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression":  Ridge(alpha=1.0),
    "Decision Tree":     DecisionTreeRegressor(max_depth=5, random_state=42)
}
print("✅ Models defined:", list(models.keys()))
```

```python
## Step 7: Model Evaluation and Comparison
##**RMSE** (Root Mean Squared Error) — lower is better
##**R² Score** — higher is better (1.0 = perfect fit, 0.0 = predicts mean only)
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    r2   = r2_score(y_test, predictions)
    results[name] = {"RMSE": round(rmse, 4), "R2 Score": round(r2, 4)}
    print(f"{name:<22} | RMSE: {rmse:.4f} | R2: {r2:.4f}")

results_df = pd.DataFrame(results).T
print("\n=== Model Comparison Table ===")
results_df
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
colors = ['#2196F3', '#4CAF50', '#FF5722']
short_names = ['Linear\nRegression', 'Ridge\nRegression', 'Decision\nTree']
rmse_vals = [results[m]['RMSE'] for m in results]
r2_vals   = [results[m]['R2 Score'] for m in results]

axes[0].bar(short_names, rmse_vals, color=colors, edgecolor='white', linewidth=1.2)
axes[0].set_title('RMSE by Model (Lower is Better)', fontweight='bold'); axes[0].set_ylabel('RMSE')
for i, v in enumerate(rmse_vals): axes[0].text(i, v+0.005, f'{v:.4f}', ha='center', fontweight='bold', fontsize=9)

axes[1].bar(short_names, r2_vals, color=colors, edgecolor='white', linewidth=1.2)
axes[1].set_title('R² Score by Model (Higher is Better)', fontweight='bold')
axes[1].set_ylabel('R² Score'); axes[1].set_ylim(0, 1.1)
for i, v in enumerate(r2_vals): axes[1].text(i, v+0.01, f'{v:.4f}', ha='center', fontweight='bold', fontsize=9)

plt.suptitle('Model Performance Comparison', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Step 8: Visual Performance Validation
##Scatter plots of Actual vs Predicted values. Points close to the red diagonal line indicate accurate predictions.
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, (name, model) in zip(axes, models.items()):
    y_pred = model.predict(X_test)
    ax.scatter(y_test, y_pred, alpha=0.3, color='steelblue', s=10)
    lim = [float(y_test.min()), float(y_test.max())]
    ax.plot(lim, lim, 'r--', linewidth=1.5, label='Perfect Prediction')
    ax.set_xlabel('Actual House Prices')
    ax.set_ylabel('Predicted House Prices')
    ax.set_title(f'{name}\nR²={results[name]["R2 Score"]:.4f}', fontweight='bold')
    ax.legend(fontsize=8)

plt.suptitle('Actual vs Predicted House Prices — All Models', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('actual_vs_predicted.png', dpi=150, bbox_inches='tight')
plt.show()
## Step 9: Best Model Selection & Feature Importance
best_model_name = max(results, key=lambda m: results[m]['R2 Score'])
print(f"🏆 Best Model  : {best_model_name}")
print(f"   RMSE        : {results[best_model_name]['RMSE']}")
print(f"   R² Score    : {results[best_model_name]['R2 Score']}")
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
feature_names = list(X.columns)

coefs = models["Linear Regression"].coef_
sorted_idx = np.argsort(np.abs(coefs))
axes[0].barh([feature_names[i] for i in sorted_idx], [coefs[i] for i in sorted_idx], color='steelblue')
axes[0].set_title('Linear Regression — Feature Coefficients', fontweight='bold')
axes[0].set_xlabel('Coefficient Value')

importances = models["Decision Tree"].feature_importances_
sorted_idx2 = np.argsort(importances)
axes[1].barh([feature_names[i] for i in sorted_idx2], [importances[i] for i in sorted_idx2], color='darkorange')
axes[1].set_title('Decision Tree — Feature Importances', fontweight='bold')
axes[1].set_xlabel('Importance Score')

plt.tight_layout()
plt.savefig('feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()
# Optional: Save the best model for deployment
import joblib
best_model = models[best_model_name]
joblib.dump(best_model, 'best_model.pkl')
print(f"✅ Best model ({best_model_name}) saved as 'best_model.pkl'")
```
